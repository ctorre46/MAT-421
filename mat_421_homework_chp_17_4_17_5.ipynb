17.4 Lagrange Polynomial Interpolation

Lagrange Polynomial Interpolation finds a single polynomial that goes through all the data points.

Ex 1:    Data set: x=[4,5,6] y=[1,7,4]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
x=[0,1,2,3]
y=[4,1,3,2]

x_new=np.arange(-0.5,5,0.1)
f=lagrange(x,y)
```

```python
fig=plt.figure
plt.plot(x_new,f(x_new),'b',x,y,'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x') # x line
plt.ylabel('y') # y line
plt.show()
```

7.5 Newton's Polynomial Interpolation

Newton's Polynomial Interpolation is a different way to fit a set of data points.

```python
def divided_diff(x, y):
    # divided differences
    n = len(y)
    coef = np.zeros([n, n])
    coef[:,0] = y

    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = \
           (coef[i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])

    return coef

def newton_poly(coef, x_data, x): #Creates a function
    n = len(x_data) - 1
    p = coef[n]
    for k in range(1,n+1):
        p = coef[n-k] + (x -x_data[n-k])*p
    return p

```

```python
x = np.array([-5, -3, -1, 1, 3])
y = np.array([4, -2, -1, 8,1])
a_s = divided_diff(x, y)[0, :] # get divided difference coef

x_new = np.arange(-5.1, 3.1, .1)
y_new = newton_poly(a_s, x, x_new)

plt.figure(figsize = (12, 8))
plt.plot(x, y, 'bo')
plt.plot(x_new, y_new)
```
