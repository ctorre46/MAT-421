**7.3 Cubic Spline Interpolation**

Cubic spline interpolation is the set of piecewise cubic functions. Cubic functions must intersect the left and right data points.

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [1, 2, 3]   # data set1
y = [3, 5, 3]   # data set2

# use bc_type = 'natural' adds the constraints as we described above
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 3, 100)  # interval
y_new = f(x_new)
```

```python
plt.figure(figsize = (20,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation') #title
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
